# Exercício 1

Faça o gráfico de valores ao lado do histograma de um vetor de valores aleatórios que segue uma distribuição de probabilidade exponencial.

```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Configura um seed aleatório para o gerador de números aleatórios
random.seed(random.random())

# Tamanho da amostra de dados
n = 5000

# Gera um conjunto de dados aleatórios de maneira exponencial
data = np.random.exponential(1, n)

# Realiza plotagem do gráfico de histograma dos dados gerados
plt.figure()
plt.subplot(1,2,1)
plt.plot(data, '.')
plt.grid()
plt.title("Gráfico de Valores")
plt.ylabel('Valores')
plt.xlabel("Frequência")
plt.subplot(1,2,2)
plt.ylabel("Frequência")
plt.xlabel("Valores")
plt.grid()
plt.title("Distribuição de Probabilidade")
plt.hist(data, bins=25)
plt.tight_layout()
plt.show()
```

# Exercício 2

Faça um programa para mostrar a curva da distribuição de probabilidade exponencial. Calcule a probabilidade de obter um valor entre
1 e 2

```python
from scipy.stats import expon
import matplotlib.pyplot as plt

# Tamanho da amostra de dados
n = 5000

# Gera dados com a biblioteca Expon com distribuição exponencial
data = np.linspace(expon.ppf(0.01), expon.ppf(0.99), n)

# Plota gráfico com a curva de distribuição exponencial
plt.figure()
plt.grid()
plt.title("Curva de Distribuição Exponencial")
plt.plot(data, expon.pdf(data),'r-', alpha=1, lw=1, label="Curva de Distribuição")

# Gera um conjunto de dados aleatórios para comparar com a distribução
random_data = expon.rvs(size=n)

# Plota o histograma dos dados aleatórios para comparar com a curva de distribuição
plt.hist(random_data, bins=25, density=True, alpha=0.5, label="Números Aleatorios")

# Calcula probabilidade entre "A" e "B"
a = 1
b = 2
prob = expon.pdf(a) - expon.pdf(b)
print(f'A probabilidade entre {a:.1f} e {b:.1f} é de {prob:.3f}')

# Plota gráfico de probabilidade

# Gera PDF com base nos dados de distribuição exponencial
exp = expon.pdf(data)

# Obtem os pontos que estão entre A e B
px = data[(data>a)&(data<b)] 

# Plota o gráficom com a probabilidade
plt.fill_between(px, exp[(data>a)&(data<b)],color='g',alpha=0.8)

plt.legend(loc="upper right")
```
