# Exercício 1

Faça o gráfico de valores ao lado do histograma de um vetor de valores aleatórios que segue uma distribuição de probabilidade exponencial.

```python
import matplotlib.pyplot as plt
import numpy as np
import random
import datetime

random.seed(datetime.datetime.now())

data = np.random.exponential(1, 5000)

plt.figure()
plt.subplot(1,2,1)
plt.plot(data, '.')
plt.grid()
plt.title("Gráfico de Valores")
plt.ylabel('Valores')
plt.xlabel("Frequência")
plt.subplot(1,2,2)
plt.ylabel("Frequência")
plt.xlabel("Valores")
plt.grid()
plt.title("Distribuição de Probabilidade")
plt.hist(data, bins=25)
plt.tight_layout()
plt.show()
```
